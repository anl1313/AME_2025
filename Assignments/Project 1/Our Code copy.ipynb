```python
#Making more variables for descriptive statistics
dat['capital'] = np.exp(dat['lcap'])
dat['ansatte'] = np.exp(dat['lemp'])
dat['dsa'] = np.exp(dat['ldsa'])
dat['empl_prod'] = dat['dsa']/dat['ansatte']
dat['cap_prod'] = dat['dsa']/dat['capital']
dat['intensity'] = dat['capital']/dat['ansatte']
```

```python
#Descriptive statistics
cols = ['lcap','lemp','ldsa',
        'capital','ansatte','dsa',
        'empl_prod','cap_prod','intensity']

def skew(x):
    # Skewness
    return stats.skew(x, bias=False, nan_policy='omit')

def exkurt(x):
    # Excess kurtosis (Fisher) => normalfordeling har 0
    return stats.kurtosis(x, fisher=True, bias=False, nan_policy='omit')

def jb_p(x):
    # Jarque-Bera test p-value
    jb = stats.jarque_bera(x, nan_policy='omit')
    return jb.pvalue

summ = pd.DataFrame({
    'Mean':   dat[cols].mean(),
    'SD':     dat[cols].std(ddof=1),
    'Min':    dat[cols].min(),
    'Max':    dat[cols].max(),
    'Skew':   dat[cols].apply(skew, axis=0),
    'ExKurt': dat[cols].apply(exkurt, axis=0),
    'JB p':   dat[cols].apply(jb_p, axis=0)
}).round(3)

print(summ)

latex = summ.to_latex(
    index=True,
    caption='Descriptive statistics: Mean, SD, Min, Max, Skewness, Excess Kurtosis (normal = 0) and Jarque–Bera p-value',
    label='tab:all_vars_desc',
    float_format="%.3f",
    escape=False
)
print(latex)
```

### FD and Serial correlation

```python
#Make FD transformation matrix
def fd_matrix(T):
    D_T = np.zeros((T-1, T))
    for i in range (T-1):
        D_T[i,i] = -1
        D_T[i, i+1] = 1
    return D_T

D_T = fd_matrix(T)

#Makes FD transformation
y_diff = est.perm(D_T, y.reshape(-1, 1))           # (N*(T-1), 1)
x_diff = est.perm(D_T, X)                           # (N*(T-1, K)

#Remove constant from x_diff
x_diff = x_diff[:, 1:]  

#Estimate FD 
fd_result = est.estimate(y_diff, x_diff, transform='fd', robust_se=True ,T=T)
label_x_fd = ['log_ansatte', 'log_kapital']

#Print results
est.print_table((label_y, label_x_fd), fd_result, title='FD regression', floatfmt='.4f')
```

```python
#Make function to calculate the serial correlation
def serial_corr(y, x, T):
    #Calculate the residuals
    if x.ndim == 1:
        x = x.reshape(-1, 1)     
    if y.ndim == 1:
        y = y.reshape(-1, 1)                    
    b_hat = la.inv(x.T @ x) @ (x.T @ y)
    e = y - x @ b_hat                         # (N*T,)

    # Create a lag transformation matrix (perm to lag)
    P_lag = np.eye(T-1, T)                    # (T-1 x T)

    #Lag residuals 
    E = e.reshape(-1, T)                      # (N, T)
    e_l = (E @ P_lag.T).reshape(-1, 1)        # (N*(T-1), 1)  ← regressor

    # Create a transformation matrix that removes the first time-period (perm to drop first)
    I_T = np.eye(T-1, T, k=1)                 # (T-1 x T)

    # Remove first observation of each individual (align "now" with the lag)
    e = (E @ I_T.T).reshape(-1,1)              # (N*(T-1),)

    #Calculate the serial correlation 
    N = e.shape[0] // (T-1)
    return est.estimate(e, e_l, transform='fd', T=T-1)
```

```python
#Estimate serial correlation from FD residuals
corr_result = serial_corr(y_diff, x_diff, T-1)
#Print results
label_ye = 'FD residual, Δe\u1d62\u209c'
label_e = ['Δe\u1d62\u209c\u208B\u2081']
est.print_table(
    (label_ye, label_e), corr_result, 
    title='Serial Correlation', floatfmt='.4f'
)
#Calculate p-value for test of no serial correlation
r = corr_result['b_hat'][0]
p_val = 2 * (1 - stats.norm.cdf(np.abs(r) / corr_result['se'][0]))
print(f'P-value for test of no serial correlation: {p_val[0]:8.4f}')
# OLS residual
corr_result = serial_corr(y, X, T)
#Print results
label_ye = 'OLS residual, Δe\u1d62\u209c'
label_e = ['e\u1d62\u209c\u208B\u2081']
est.print_table(
    (label_ye, label_e), corr_result, 
    title='Serial Correlation', floatfmt='.4f'
)
#Calculate p-value for test of no serial correlation
r = corr_result['b_hat'][0]
p_val = 2 * (1 - stats.norm.cdf(np.abs(r) / corr_result['se'][0]))
print(f'P-value for test of no serial correlation: {p_val[0]:8.4f}')
```

### Test for strict exogeneity

```python
# Create leads for both employment and capital
F_T = np.eye(T, T, k=1)  # Shift one period forward

# Create leads for both variables
emp_lead_full = est.perm(F_T, X[:, 1].reshape(-1, 1))  # Employment lead (N*T, 1)
cap_lead_full = est.perm(F_T, X[:, 2].reshape(-1, 1))  # Capital lead (N*T, 1)

# Remove the last observed year for every individual
I_T = np.eye(T-1, T)

x_exo = est.perm(I_T, X)                           # (N*(T-1), K)
y_exo = est.perm(I_T, y.reshape(-1, 1))            # (N*(T-1), 1)
emp_lead = est.perm(I_T, emp_lead_full)            # (N*(T-1), 1)
cap_lead = est.perm(I_T, cap_lead_full)            # (N*(T-1), 1)

# Check that the dimensions are correct
assert x_exo.shape[0] == emp_lead.shape[0] == cap_lead.shape[0] == y_exo.shape[0]

# Add both leads to x_exo
x_exo = np.hstack((x_exo, emp_lead, cap_lead))     # (N*(T-1), K+2)

# Within transform the data (demean over T-1)
Tm1 = T - 1
Q_T = np.eye(Tm1) - np.ones((Tm1, Tm1)) / Tm1      # (T-1)×(T-1)

# Apply panel-wise with perm helper
yw_exo = est.perm(Q_T, y_exo.reshape(-1, 1)).reshape(-1,)   # (N*(T-1),)
xw_exo = est.perm(Q_T, x_exo)                               # (N*(T-1), K+2)

# Select variables: employment, capital, emp_lead, cap_lead
xw_exo = np.hstack((xw_exo[:, 1:3], xw_exo[:, -2:]))  # (N*(T-1), 4)

# Estimate model
yw_exo = yw_exo.reshape(-1, 1)
exo_test = est.estimate(yw_exo, xw_exo, T=T-1, robust_se=True, transform='fe')

# Print results
label_exo = ['log_ansatte', 'log_kapital', 'log_ansatte_t+1', 'log_kapital_t+1']
est.print_table((label_y, label_exo), exo_test, title='Strict Exogeneity Test', floatfmt='.4f')

# Joint Wald test: H0: both leads = 0 (strict exogeneity holds)
b = exo_test['b_hat']
cov = exo_test['cov']

# Test matrix for both leads
R_leads = np.zeros((2, len(b)))
R_leads[0, 2] = 1  # emp_lead coefficient
R_leads[1, 3] = 1  # cap_lead coefficient

# Wald statistic
wald_stat = (R_leads @ b).T @ la.inv(R_leads @ cov @ R_leads.T) @ (R_leads @ b)
p_value = stats.chi2.sf(float(wald_stat), df=2)

print(f'\nJoint test H0: Both leads = 0 (strict exogeneity)')
print(f'Wald statistic: {float(wald_stat):.4f}')
print(f'P-value: {p_value:.4f}')

# Individual t-tests for each lead
print(f'\nIndividual tests:')
for i, name in enumerate(['Employment lead', 'Capital lead']):
    t_stat = b[i+2] / np.sqrt(cov[i+2, i+2])
    p_val = 2 * (1 - stats.norm.cdf(np.abs(t_stat)))
    print(f'{name}: t = {float(t_stat):.4f}, p-value = {float(p_val):.4f}')  
```

### Augmented FD Test (Exogeneity)

This test augments the FD regression with:
- **Lagged levels** ($x_{it}$): Tests sequential exogeneity / dynamic specification
- **Leads** ($x_{i,t+1}$): Tests strict exogeneity

Interpretation:
- If **levels** significant: Sequential exogeneity may fail 
- If **leads** significant: Strict exogeneity fails (future $x$ predicts current $\Delta y$)
- Both tests together help distinguish weak vs. strict exogeneity

```python
# Need to align data: drop first period for levels, drop last period for leads
# Work with T-2 observations to have both x_t and x_{t+1} available

# Create transformation matrices
R_T = np.eye(T-2, T-1, k=0)   # (T-2) x (T-1), keeps periods 2 to T-1 from y_diff
L_T = np.eye(T-2, T, k=1)     # (T-2) x T, selects x at periods 2 to T-1 (levels)
F_T = np.eye(T-2, T, k=2)     # (T-2) x T, selects x at periods 3 to T (leads)

# Apply transformations
y_diff_trunc = est.perm(R_T, y_diff)           # (N*(T-2), 1)
x_diff_trunc = est.perm(R_T, x_diff)           # (N*(T-2), K-1)
x_level = est.perm(L_T, X)[:, 1:]              # (N*(T-2), K-1), drop constant
x_lead = est.perm(F_T, X)[:, 1:]               # (N*(T-2), K-1), drop constant

# Verify dimensions match
assert x_diff_trunc.shape[1] == x_level.shape[1] == x_lead.shape[1], \
    f"Column mismatch: diff={x_diff_trunc.shape[1]}, level={x_level.shape[1]}, lead={x_lead.shape[1]}"

# Assemble augmented RHS: [Δx, x_t, x_{t+1}]
X_fd_full = np.hstack([x_diff_trunc, x_level, x_lead])

labels_full = ['Δlog_ansatte', 'Δlog_kapital', 
               'log_ansatte_t', 'log_kapital_t',
               'log_ansatte_t+1', 'log_kapital_t+1']

# Estimate full augmented FD regression
fd_full_res = est.estimate(y_diff_trunc, X_fd_full, transform='fd', T=T-2, robust_se=True)
est.print_table((label_y, labels_full), fd_full_res, title='Augmented FD: Levels + Leads', floatfmt='.4f')

# Joint tests
b = fd_full_res['b_hat']
cov = fd_full_res['cov']

# Test 1: H0: levels not significant (sequential exogeneity)
R_levels = np.zeros((2, len(b)))
R_levels[0, 2] = 1  # log_ansatte_t
R_levels[1, 3] = 1  # log_kapital_t

wald_levels = (R_levels @ b).T @ la.inv(R_levels @ cov @ R_levels.T) @ (R_levels @ b)
p_levels = stats.chi2.sf(float(wald_levels), df=2)

print(f'\nTest 1: H0: Levels not significant (sequential exogeneity holds)')
print(f'Wald statistic: {float(wald_levels):.4f}, P-value: {p_levels:.4f}')

# Test 2: H0: leads not significant (strict exogeneity)
R_leads = np.zeros((2, len(b)))
R_leads[0, 4] = 1  # log_ansatte_t+1
R_leads[1, 5] = 1  # log_kapital_t+1

wald_leads = (R_leads @ b).T @ la.inv(R_leads @ cov @ R_leads.T) @ (R_leads @ b)
p_leads = stats.chi2.sf(float(wald_leads), df=2)

print(f'\nTest 2: H0: Leads not significant (strict exogeneity holds)')
print(f'Wald statistic: {float(wald_leads):.4f}, P-value: {p_leads:.4f}')
```

### Test for CRS

```python
######################POLS##########################
b_hat = pooled_result['b_hat']
cov   = pooled_result['cov']

ansatte_index = label_x.index('log_ansatte')
kapital_index = label_x.index('log_kapital')

b = np.asarray(b_hat).reshape(-1)   # (P,)
V = np.asarray(cov)                 # (P,P)
P = b.size
if V.shape != (P, P):
    raise ValueError(f"Kovarians har forkert form: {V.shape}, forventede {(P,P)}")

R = np.zeros((1, P))
R[0, [ansatte_index, kapital_index]] = 1.0

#Wald-test for H0: beta_kapital + beta_ansatte = 1
num = float(R @ b - 1.0)           # c'β - q, q=1
den = float(R @ V @ R.T)           # c' V c
if not np.isfinite(den) or den <= 0:
    raise ValueError("Check")

wald = (num ** 2) / den
p_value = stats.chi2.sf(wald, df=1)

print(f"Wald test statistic: {wald:.4f}")
print(f"p-value: {p_value:.4g}")
```
